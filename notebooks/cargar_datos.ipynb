```python
# Filtramos edades atípicas (errores de carga)
df = df[df["edad_cliente"] <= 100]

df["edad_cliente"].max(), df.shape
```

Se detectaron valores atípicos en la variable edad_cliente (122 y 123 años), los cuales se consideraron errores de carga. Se decidió filtrar registros con edades mayores a 100 años para mantener coherencia en el análisis.
